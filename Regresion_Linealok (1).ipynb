```python
# Importa las bibliotecas necesarias

import pandas as pd
import numpy as np
import os

from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Definición de variable y cargue de archivo data
lineal = pd.read_csv("C:/Users/LILIANA CARDONA/OneDrive/Escritorio/INGENIERIA SISTEMAS/C/6. Periodo 2023 II PERIODO 16-04 (1394)/2. ANÁLISIS DE DATOS/ETAPA 2/archive RL/data.csv")
```

```python
df = pd.DataFrame(lineal)
```

```python
# Define las variables independientes (X) y dependientes (Y)
X = df[['metro']]
Y = df['precio']
```

```python
# Crea una instancia del modelo de regresión lineal
modelo_regresion = LinearRegression()
```

```python
# Ajusta el modelo a los datos

modelo_regresion.fit(X, Y)
```

```python
# Ajusta el modelo a tus datos

modelo_regresion.fit (X, Y)
```

```python
# Imprime los coeficientes del modelo

print('Intersección (beta0):', modelo_regresion.intercept_)
print('Pendiente (beta1):', modelo_regresion.coef_[0])
```

```python
# Realiza predicciones
predicciones = modelo_regresion.predict(X)
```

```python
# Grafica los datos y la línea de regresión

plt.scatter(X, Y, label='Datos reales')
plt.plot(X, predicciones, color='red', label='Línea de regresión')
plt.xlabel('Metro')
plt.ylabel('Precio')
plt.legend()
plt.show()
```
